### a4 vs un

```python
gff = pd.read_csv("../chipseeker_removeCM/TAIR10_GFF3_genes.gff", sep="\t", header=None)
gff = gff.loc[gff.iloc[:,2] == "gene", :]

# ### a4 vs un
# origin_peak = pd.read_csv("wt_a4_vs_un_peaks.narrowPeak", sep="\t", header=None)
# a4_peak = pd.read_csv("wt_a4_vs_un_peaks_annotation.txt", sep=" ")

# annotated_peak = np.unique(a4_peak.peak.str[1:].astype(int)-1)
# not_annotated_peak = [i for i in origin_peak.index.values if i not in annotated_peak]
# not_annotated_peak = origin_peak.iloc[not_annotated_peak, :]
# not_annotated_peak.head()

def get_dACR(not_annotated_peak, gff):
    dACR = []
    for i in range(not_annotated_peak.shape[0]):
        chrom = not_annotated_peak.iloc[i, 0]
        start = not_annotated_peak.iloc[i, 1]
        end = not_annotated_peak.iloc[i, 2]

        if gff[(gff.iloc[:, 0] == chrom) & (gff.iloc[:, 3] <= start) & (gff.iloc[:, 4] >= end)].shape[0] > 0:
            pass
        else:
            tmp_gff1 = gff[(gff.iloc[:, 0] == chrom) & (gff.iloc[:, 4] <= start)]
            tmp_gff2 = gff[(gff.iloc[:, 0] == chrom) & (gff.iloc[:, 3] >= end)]
            if tmp_gff1.shape[0] < 1:
                if tmp_gff2.iloc[0, 6] == "+":
                    dACR.append([chrom, start, end, tmp_gff2.iloc[0, 3] - end, tmp_gff2.iloc[0, 8][3:12]])
            elif tmp_gff2.shape[0] < 1:
                if tmp_gff1.iloc[-1, 6] == "-":
                    dACR.append([chrom, start, end, start - tmp_gff1.iloc[-1, 4], tmp_gff1.iloc[-1, 8][3:12]])
            else:
                if tmp_gff1.iloc[-1, 6] == "+":
                    if tmp_gff2.iloc[0, 6] == "-":
                        pass
                    else:
                        dACR.append([chrom, start, end, tmp_gff2.iloc[0, 3] - end, tmp_gff2.iloc[0, 8][3:12]])
                else:
                    if tmp_gff2.iloc[0, 6] == "-":
                        dACR.append([chrom, start, end, start - tmp_gff1.iloc[-1, 4], tmp_gff1.iloc[-1, 8][3:12]])
                    else:
                        ## shared dACR peak
                        pass
    dACR = pd.DataFrame(dACR)
    dACR.columns = ["chr", "start", "end", "distance", "target"]
    return dACR

# dACR = get_dACR(not_annotated_peak, gff)
# dACR.head()
```

```python
### a4 vs un
origin_peak = pd.read_csv("wt_a4_vs_un_peaks.narrowPeak", sep="\t", header=None)
a4_peak = pd.read_csv("wt_a4_vs_un_peaks_annotation.txt", sep=" ")

annotated_peak = np.unique(a4_peak.peak.str[1:].astype(int)-1)
not_annotated_peak = [i for i in origin_peak.index.values if i not in annotated_peak]
not_annotated_peak = origin_peak.iloc[not_annotated_peak, :]
a4_dACR = get_dACR(not_annotated_peak, gff)

sum_ATAC_targets = list(a4_peak.feature)
sum_ATAC_targets.extend(list(a4_dACR.target))
sum_ATAC_targets = np.unique(sum_ATAC_targets)

DEG = pd.read_csv("../DEG/Significant DE genes list and statistics.csv")
a4_vs_un = DEG[DEG["contrast"] == "wt.a4-wt.un"]
a4_vs_un = a4_vs_un[a4_vs_un["up.down"] == "up-regulated"]

peak_specific = len(sum_ATAC_targets) - len(list(set(sum_ATAC_targets) & set(a4_vs_un.target)))
DEG_specific = len(np.unique(a4_vs_un.target)) - len(list(set(sum_ATAC_targets) & set(a4_vs_un.target)))
a4_common = list(set(sum_ATAC_targets) & set(a4_vs_un.target))
venn2(subsets = (peak_specific, DEG_specific, len(a4_common)), set_labels = ("ATACpeak_target\nspecific_in_a4(vs_un)", "upregulated_in_a4(vs_un)"))
```

## Correlation between peak fold change(MACS2 output) & TPM Log2fold change a4 vs un
### a4 vs un

```python
### a4 vs un
origin_peak = pd.read_csv("wt_a4_vs_un_peaks.narrowPeak", sep="\t", header=None)
a4_peak = pd.read_csv("wt_a4_vs_un_peaks_annotation.txt", sep=" ")

peak_scores = []
for i in range(a4_peak.shape[0]):
    peak_score = origin_peak.iloc[int(a4_peak.iloc[i, 5][1:])-1, 4]
    peak_scores.append(peak_score)

a4_peak["score"] = peak_scores

unique_target_and_score = []
for i in np.unique(a4_peak.feature.values):
    unique_target_and_score.append([i, a4_peak[a4_peak["feature"] == i].score.max()])
unique_target_and_score = pd.DataFrame(unique_target_and_score)
unique_target_and_score.columns = ["target", "score"]
unique_target_and_score.head()
```

```python
DEG = pd.read_csv("../DEG/Significant DE genes list and statistics.csv")
a4_vs_un = DEG[DEG["contrast"] == "wt.a4-wt.un"]
a4_vs_un = a4_vs_un[a4_vs_un["up.down"] == "up-regulated"]

common = list(set(unique_target_and_score.target) & set(a4_vs_un.target))
print(len(common))
scores = unique_target_and_score[unique_target_and_score.target.isin(common)].score
logFC2 = a4_vs_un[a4_vs_un.target.isin(common)].log2FC
print(len(scores), len(logFC2))
plt.scatter(scores, logFC2)

correlation_matrix = np.corrcoef(scores, logFC2)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
r_squared
```

```python
pd.DataFrame({"target":list(common),
              "scores":list(scores),
              "logFC2":list(logFC2)}).sort_values(by=(["scores"]), ascending=False).head()
```
